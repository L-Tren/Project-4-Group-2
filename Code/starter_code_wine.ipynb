```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load 
red_wine = Path("Resources/winequality-red.csv")
white_wine = Path("Resources/winequality-white.csv")

# Read Data File and store into Pandas DataFrames
redwine_data = pd.read_csv(red_wine,delimiter=';')
whitewine_data = pd.read_csv(white_wine,delimiter=';')

```

```python
# Add Type column to ditinguish between red and white wine
redwine_data.insert(0, column = "Type", value = "Red")
redwine_data['Wine Quality Categorisation'] = redwine_data['quality'].apply(lambda x: 1 if x >= 6 else 0)
redwine_data.head()
```

```python
# Add Type column to ditinguish between red and white wine
whitewine_data.insert(0, column = "Type", value = "White")
whitewine_data['Wine Quality Categorisation'] = whitewine_data['quality'].apply(lambda x: 1 if x >= 6 else 0)
whitewine_data.head()
```

```python
#Drop column quality as its replaced by Wine Quality Categorisation
redwine_dropped = redwine_data.drop(columns=['quality'])
redwine_dropped.head()
```

```python
##Drop column quality as its replaced by Wine Quality Categorisation
whitewine_dropped = whitewine_data.drop(columns=['quality'])
whitewine_dropped.head()
```

```python
# save combined data to csv
whitewine_dropped.to_csv('whitewine_quality.csv')
redwine_dropped.to_csv('redwine_quality.csv')
```
